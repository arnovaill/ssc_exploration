```python
import sys
sys.path.append("../SSC")
import argparse
import torch
import torch.nn as nn
import torch.utils.data
from torch.autograd import Variable
import numpy as np
import os
import glob
import datetime
from tqdm import tqdm
import models
import time 
import matplotlib.pyplot as plt
import mpl_toolkits
import matplotlib.patches as mpatches
from mpl_toolkits.mplot3d import Axes3D
from models import make_model
import pickle
print(torch.__version__)

def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

# ---- Check CUDA
if torch.cuda.is_available():
    print("Great, You have {} CUDA device!".format(torch.cuda.device_count()))
else:
    print("Sorry, You DO NOT have a CUDA device!")

    



colorMap = np.array([[22, 191, 206],    # 0 empty, free space
                     [214,  38, 40],    # 1 ceiling
                     [43, 160, 4],      # 2 floor
                     [158, 216, 229],   # 3 wall
                     [114, 158, 206],   # 4 window
                     [204, 204, 91],    # 5 chair  new: 180, 220, 90
                     [255, 186, 119],   # 6 bed
                     [147, 102, 188],   # 7 sofa
                     [30, 119, 181],    # 8 table
                     [188, 188, 33],    # 9 tvs
                     [255, 127, 12],    # 10 furn
                     [196, 175, 214],   # 11 objects
                     [153, 153, 153],     # 12 label==255, ignore
                     ]).astype(np.int32)

def rgb2hex(r,g,b):
    return "#{:02x}{:02x}{:02x}".format(r,g,b)

colorMapHex = [rgb2hex(color[0],color[1],color[2]) for color in colorMap]
labels = ['empty','ceiling','floor','wall','window','chair','bed','sofa','table','tvs','furn','objects','12 label==255, ignore']

def create_legend(labels, colors):
    """
    Creates a legend for plots using the given colors and labels.

    :param labels: Iterable of labels that shall be used in the legend.
    :param colors: Iterable of colors that shall be used in the legend.
    :return:
    """
    patches = []

    for color, label in zip(colors, labels):
        patches.append(mpatches.Patch(color=color, label=label))

    plt.legend(handles=patches, prop={'size': 12})
```

```python
names = glob.glob('/home/arno/Documents/NYUCAD_LATEST/NYUCADtest_npz/*.npz')
_name = names[72]

npz_file = np.load(_name) 
print(npz_file.keys())
tsdf_hr = npz_file['tsdf_hr']
tsdf_lr = npz_file['tsdf_lr']
position = npz_file['position']
target_lr = npz_file['target_lr']
depth = npz_file['depth']
rgb = npz_file['rgb']
rgb.shape
```

```python
#torch.backends.cudnn.enabled = True
torch.backends.cudnn.benchnark=True
net = make_model('ddrnet', num_classes=12).cuda().eval()
count_parameters(net)
```

```python
load_checkpoint = torch.load('/home/arno/SSC/weights/cpBest_SSC_DDRNet3.pth.tar')

state_dict = dict()
for key in load_checkpoint['state_dict'].keys():
    new_key = key.replace('module.','')
    state_dict[new_key] = load_checkpoint['state_dict'][key]

net.load_state_dict(state_dict)
```

```python
var_depth = Variable(torch.unsqueeze(torch.tensor(depth),0).float()).cuda()
var_rgb = Variable(torch.unsqueeze(torch.tensor(rgb),0).float()).cuda()
position = torch.tensor(position).long().cuda()
```

```python
print(torch.sum(torch.isnan(var_depth)),torch.sum(torch.isnan(var_rgb)),torch.sum(torch.isnan(position))) # check for nan 
print(torch.max(var_depth),torch.max(var_rgb),torch.max(position))
print(torch.min(var_depth),torch.min(var_rgb),torch.min(position))
```

```python
with open('/home/arno/SSC/data_unreal/var_depth.pickle', 'rb') as f:
    var_depth = pickle.load(f)
with open('/home/arno/SSC/data_unreal/var_rgb.pickle', 'rb') as f:
    var_rgb = pickle.load(f)
with open('/home/arno/SSC/data_unreal/position.pickle', 'rb') as f:
    position = pickle.load(f)
```

```python
depth = var_depth.cpu().numpy()
super_position = position.cpu().numpy()
depth_0 = np.where(super_position==0,0,depth)
var_depth = Variable(torch.tensor(depth_0).float()).cuda()
```

```python
print(torch.sum(torch.isnan(var_depth)),torch.sum(torch.isnan(var_rgb)),torch.sum(torch.isnan(position))) # check for nan 
print(torch.max(var_depth),torch.max(var_rgb),torch.max(position))
print(torch.min(var_depth),torch.min(var_rgb),torch.min(position))
```

```python
# start timer 

for i in range(5):
    torch.cuda.synchronize()
    start_t1 = time.clock() 

    with torch.no_grad():
        pred = net(x_depth = var_depth, x_rgb=var_rgb, p = position)
    torch.cuda.synchronize()
    print('t1:',time.clock()-start_t1)
    #torch.cuda.synchronize()
    
    start_time = time.clock() 
    y_pred = pred.cpu().data.numpy()  # CUDA to CPU, Variable to numpy
    print('t2:',time.clock()-start_time)
    predict_completion = np.squeeze(np.argmax(y_pred, axis=1))
    #print('t_tot = ',time.clock()-start_t1)
```

```python
#%matplotlib qt
%matplotlib inline
# MIRROR PROBLEM 
#target_clean = np.flip(target_lr,axis= 0) 
# Change order for visualization
target_clean = np.swapaxes(target_lr,1,2)
print(target_clean.shape)



target_clean[target_clean ==255] = 0
np.max(target_clean)
voxels_occ = target_clean != 0

# set the colors of each object
colors = np.empty(voxels_occ.shape, dtype=object)

colormap = colorMap[target_clean]

# Safety verif
u = np.reshape(np.reshape(colormap,(60*36*60,3)),(60,60,36,3))
if np.sum(u != colormap) > 0: 
    print('PROBLEM')

colormap_flat = np.reshape(colormap,(60*36*60,3))
colormap_hex_flat = [rgb2hex(color[0],color[1],color[2]) for color in colormap_flat]
colormap_hex = np.reshape(colormap_hex_flat,(60,60,36))

# and plot everything
fig = plt.figure(figsize=(20,10))
ax = fig.gca(projection='3d')
ax.voxels(voxels_occ, facecolors=colormap_hex, edgecolor='k')
create_legend(labels[1:-1], colorMapHex[1:-1])

# ‘elev’ stores the elevation angle in the z plane. ‘azim’ stores the azimuth angle in the x,y plane.
ax.view_init(elev=20., azim=-120) 
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
